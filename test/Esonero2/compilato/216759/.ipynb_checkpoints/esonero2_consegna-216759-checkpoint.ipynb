```python
import cv2
import numpy as np
import imutils
import matplotlib.pyplot as plt
from utils import utility as uti
import scipy.ndimage as ndimage
import os
directory=r'D:\analisi_foto_video'

import torch

###ESERCIZIO1

#####utilizziamo la piramide gaussiana per introdurre il concetto di scale_space, di fatto l'applicazione del filtro simula la perdita di dettagli che ci sarebbe con una scala ridotta



#introduciamo quindi il concetto di scale invariant, infatti dimostriamo che le feature importanti rimangono anche con una scala ridotta
image=cv2.imread(os.path.join(directory,"figure1.jpg"))

image=image[:,:,1]
rows,cols = image.shape

output = np.zeros((rows,cols+cols//2))

output[:rows,:cols] = image

num_octave = 6

cur_scale = image

horiz = True
rows = 0

for i in range(num_octave):
    sigma = 2**i
    cur_scale = ndimage.gaussian_filter(image,sigma)
    cur_scale = cur_scale[::2**(i+1),::2**(i+1)]
    rows_scaled, cols_scaled = cur_scale.shape
    output[rows:rows+rows_scaled,cols:cols+cols_scaled] = cur_scale
    rows += rows_scaled


plt.figure(figsize=(20, 20))

plt.imshow(output,cmap="gray")
plt.axis('off')
plt.show()



sz = 4


cur_scale = np.copy(image)
sigma = 2

for i in range(num_octave):
    cur_scale = ndimage.gaussian_filter(cur_scale, sigma)
    cur_scale = cur_scale[::2, ::2]

    plt.figure(figsize=(sz, sz))
    plt.imshow(cur_scale, cmap="gray")
    s = 2 * np.ceil(3 * 2 ** (i - 1)) + 1
    plt.title(f'$\sigma$: {2 ** (i - 1)} - Size: {cur_scale.shape}')
    plt.axis('off')
    plt.show()

    sz = sz / 2


```

```python
import torchvision
from torchvision.utils import make_grid

conv1 = torch.nn.Conv2d(3, 10, kernel_size=3, padding=1, stride=2)
conv2 = torch.nn.Conv2d(10, 30, kernel_size=(3, 5), padding=(0, 1), stride=(3, 4))

model = lambda x: conv2(conv1(x))

image=cv2.imread(os.path.join(directory,"figure1.jpg"))

image=cv2.resize(image,(400,400))/255


img_tensor=torch.tensor(np.reshape(image,(1,3,400,400)),dtype=torch.float)
risultato=model(img_tensor)
features=risultato[:].permute(1,0,2,3)
img_res=make_grid(features,padding=5,normalize=True).permute(1,2,0).detach().numpy()
plt.figure(figsize=(30,30))
plt.subplot(1,1,1)
#plt.imshow(img_tensor,cmap="gray",interpolation="none")

# edit by mail 17 maggio 2021 11:42:25 CEST
plt.imshow(img_res,cmap="gray",interpolation="none")
```
